```python
import pandas
from IPython.display import display
```

```python
df = pandas.read_json("data-inter/cleaned.json")
pandas.set_option("display.max_rows", 128)
pandas.set_option("display.min_rows", 64)
```

```python
display(df)
```

```python
df.groupby("backend").count()
```

```python
df_optix = df[df["backend"] == "OPTIX"] \
    .groupby(["scene", "device"]) \
    .agg(count=("timestamp", "count"), time_nosync_mean=("time_nosync", "mean"), time_total=("time_total", "mean")) \
    .sort_values(["scene", "count"], ascending=False)
```

```python
df_optix.to_csv("out.csv")
```

```python
df_optix[df_optix["count"] > 1].sort_values(["scene", "time_total"], ascending=True).to_csv("out2.csv")
```
